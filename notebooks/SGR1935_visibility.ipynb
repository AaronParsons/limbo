```python
import limbo.telescope as telescope

import numpy as np
import matplotlib.pyplot as plt
import time
from astropy.coordinates import SkyCoord, EarthLocation, AltAz
from astropy.time import Time
import astropy.units as u

%matplotlib inline

plt.rcParams['mathtext.fontset'], plt.rcParams['font.family'] = 'stix', 'STIXGeneral'
plt.rcParams.update({'font.size': 14})
```

```python
# Leuschner coordinates
LAT, LON, HEIGHT = telescope.LAT, telescope.LON, telescope.HEIGHT
leuschner = EarthLocation(lat=LAT, lon=LON, height=HEIGHT)
```

```python
sgr = SkyCoord(telescope.SGR_RA, telescope.SGR_DEC, frame='icrs')
```

```python
# utcoffset = -8*u.hour  # PTS
midnight = Time('2012-02-13 00:00:00') # UTC
delta_midnight = np.linspace(0, 24, 1000)*u.hour
altaz_frame = AltAz(obstime=midnight+delta_midnight, location=leuschner)
sgraltaz = sgr.transform_to(altaz_frame)
```

```python
alts, azs = sgraltaz.alt.deg, sgraltaz.az.deg
```

```python
fig, axes = plt.subplots(2, 1, figsize=(7,6), sharex=True, constrained_layout=True)
axes[0].set_title('Visibility of SGR1935+2154\nover Leuschner Observatory')

axes[0].plot(delta_midnight, alts, color='black', label='SGR1935 path')
axes[0].set_ylabel('Altitude')
axes[0].grid()
axes[0].hlines(telescope.ALT_MIN, min(delta_midnight.value), max(delta_midnight.value), 'red', linestyle='dashed', label='LRO alt limits')
axes[0].hlines(telescope.ALT_MAX, min(delta_midnight.value), max(delta_midnight.value), 'red', linestyle='dashed')
# axes[0].legend()

axes[1].plot(delta_midnight, azs, color='black')
axes[1].set_ylabel('Azimuth')
axes[1].set_xlabel('Hours since midnight (PTS) 02-12-2023')
axes[0].grid()
axes[1].hlines(telescope.AZ_MIN, min(delta_midnight.value), max(delta_midnight.value), 'red', linestyle='dashed')
axes[1].hlines(telescope.AZ_MAX, min(delta_midnight.value), max(delta_midnight.value), 'red', linestyle='dashed')

plt.show()
```
